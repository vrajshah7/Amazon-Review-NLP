```python
cookie = search_response.cookies
def Searchreviews(review_link):
    url="https://www.amazon.in"+review_link
    print(url)
    page=requests.get(url,cookies=cookie,headers=header)
    if page.status_code==200:
        return page
    else:
        return "Error"
```

```python
reviews = []
for k in range(82):
        response=Searchreviews('/product-reviews/0553175211/ref=cm_cr_getr_d_paging_btm_next_2?ie=UTF8&showViewpoints=1&pageNumber='+str(k))
        soup=BeautifulSoup(response.content)
        for i in soup.findAll("span",{'data-hook':"review-body"}):
            reviews.append(i.text)
```

```python
def preprocess(text):
    clean_data = []
    for x in (text[:]): 
        new_text = re.sub('<.*?>', '', x)   # remove HTML tags
        new_text = re.sub(r'[^\w\s]', '', new_text) # remove punc.
        new_text = re.sub(r'\d+','',new_text)# remove numbers
        new_text = new_text.lower() # lower case, .upper() for upper          
        if new_text != '':
            clean_data.append(new_text)
    return clean_data
```

```python
Rev_list = Df_pd["reviews"]
```

```python
def tokenization_w(words):
    w_new = []
    for w in (words[:]):
        w_token = word_tokenize(w)
        if w_token != '':
            w_new.append(w_token)
    return w_new
```

```python
lemmatizer = WordNetLemmatizer()
def lemmatization(words):
    new = []
    for i in range(len(reviews)):
        lem_words = [lemmatizer.lemmatize(x) for x in (words[:][i])]
        new.append(lem_words)
    return new
```

```python
test = ['\nYou like to eat apples. He has eaten many apples because he likes eating.\n']
test_pd = pd.DataFrame(test) #makes this into a panda data frame
clean_test = preprocess(test_pd) #removes punctuation, see above
clean_words = tokenization_w(clean_test) # word tokenization
lemtest = lemmatization(clean_words)
lemtest
```
